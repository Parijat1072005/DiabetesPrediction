```python
import pandas as pd
url = 'https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv'
df = pd.read_csv(url)
df.head()
```

```python
import numpy as np
np.mean(df['Insulin'])
```

```python
df.isnull().sum()
```

```python
mean_val_1 = df['Insulin'][df['Insulin'] != 0].mean()
df['Insulin'].replace(0, mean_val_1, inplace = True)
mean_val_2 = df['SkinThickness'][df['SkinThickness'] != 0].mean()
df['SkinThickness'].replace(0, mean_val_2, inplace = True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[: , :8]
y = df.iloc[: , -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = (.1), random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf =  LogisticRegression()
```

```python
clf.fit(x_train, y_train)
```

```python
pred_val = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(pred_val, y_test)
```

```python
import pickle
```

```python
pickle.dump(clf, open('model.pkl', 'wb'))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
training_accuracy_list = []
test_accuracy_list = []
```

```python
for n_neighbors in range (1,11):
    knn = KNeighborsClassifier(n_neighbors = n_neighbors)
    knn.fit(x_train, y_train)
    training_accuracy_list.append(knn.score(x_train, y_train))
    test_accuracy_list.append(knn.score(x_test, y_test))
```

```python
print('training accuracy is:', training_accuracy_list)
print('test accuracy is:', test_accuracy_list)
```
